#### Abstract
Parkinson's diseases is progressive central nervous system condition that affects movements and causes tremors and stiffness. 
It has 5 stages to it and an estimated 7 to 10 million people worldwidehave Parkinson's diseases.
This is chronic and still has no cure. It is a neurodegenrative disease that affects the neurons in the brain that conatin dopamine.   

Number of Instances: 195

Number of Attributes: 24

File size: 40 KB

#### Attribute Information:
name - ASCII subject name and recording number (ID)
MDVP:Fo(Hz) - Average vocal fundamental frequency
MDVP:Fhi(Hz) - Maximum vocal fundamental frequency
MDVP:Flo(Hz) - Miniimum vocal fundamental frequency
MDVP:Jitter(%), MDVP:Jitter(Abs), MDVP:RAP, MDVP:PPQ, Jitter:DDP - Several measures of variation in fundamental frequency 
MDVP:Shimmer, MDVP:Shimmer(dB), Shimmer:APQ3, Shimmer:APQ5, MDVP:APQ, Shimmer:DDA - Several measures of variation in amplitude	
NHR, HNR - Two measures of ratio of noise to tonal components in the voice
status - Health status of the subject 1 -> Parkinson's, 0 -> healthy
RPDE, D2 - Two non-linear dynamically complexity measures  	
DFA - Signal fractal scaling component	
spread1	spread2, PPE - Three non-linear measures of fundamental frequency varitaion.

```python
X = dff.drop(['status'], axis=1)
y = dff['status']
```

#### Featuring Engineering

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_ros, y_ros = ros.fit_resample(X,y)
print(y.value_counts())
print("**********************")
print(y_ros.value_counts())
```

#### Feature Scaling            

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
X = scaler.fit_transform(X_ros)
y = y_ros
```

*Applying PCA*

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
print("Before PCA", X.shape)
print("After PCA", X_pca.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, y, test_size = 0.2, random_state = 0)
```

#### Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
list_met = []
list_accuracy = []

#Model 1 - logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
# prediction
lr_pred_train = lr.predict(X_train)
lr_pred_test = lr.predict(X_test)
#accuracy
accuracy_LR = accuracy_score(Y_test, lr_pred_test)

#Model 2 - Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, Y_train) 
#prediction
dt_pred_train = dt.predict(X_train)
dt_pred_test = dt.predict(X_test)
# accuracy
accuracy_DT = accuracy_score(Y_test, dt_pred_test)

#Model 3 - RandomForest ginni
from sklearn.ensemble import RandomForestClassifier
rfg = RandomForestClassifier(random_state = 14)
rfg.fit(X_train, Y_train)
#predict
rfg_pred_train = rfg.predict(X_train)
rfg_pred_test = rfg.predict(X_test)
#evalutaion
accuracy_RFI = accuracy_score(Y_test, rfg_pred_test)

#Model 4 - random Forest entropy
rfe = RandomForestClassifier(criterion = 'entropy', random_state = 14)
rfe.fit(X_train, Y_train)
#predict
rfe_pred_train = rfe.predict(X_train)
rfe_pred_test = rfe.predict(X_test)
#evalutaion
accuracy_RFE = accuracy_score(Y_test, rfe_pred_test)

#Model 5 - SupportVectorMachine
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, Y_train)
#predict
svm_pred_train = svm.predict(X_train)
svm_pred_test = svm.predict(X_test)
#evalutaion
accuracy_SVC = accuracy_score(Y_test, svm_pred_test)

#Model 6 - KNN 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
#prediction
knn_pred_train = knn.predict(X_train)
knn_pred_test = knn.predict(X_test)
#evaluation
accuracy_knn = accuracy_score(Y_test, knn_pred_test)

list1 = ["Logistic Regression", "Decision Tree", "RandomForest Ginni", "RandomForest Entropy", 
         "Support Vector", "K Nearest Neighors",]

list2 = [accuracy_LR, accuracy_DT, accuracy_RFI, accuracy_RFE, accuracy_SVC, accuracy_knn]

df_accuracy = pd.DataFrame({'Method Used': list1, "Accuracy": list2})
print(df_accuracy)

charts = sns.barplot(x = 'Method Used', y = 'Accuracy', data = df_accuracy)
charts.set_xticklabels(charts.get_xticklabels(), rotation = 90)
```
